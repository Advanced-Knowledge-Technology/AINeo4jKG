````python
prompt_folder_path = "/content/drive/MyDrive/2_CNTTNC/TEXT2GRAPH/prompt_1410"
structure_data_folder_path = "/content/drive/MyDrive/2_CNTTNC/TEXT2GRAPH/structure_data"
csv_result_path = "/content/drive/MyDrive/2_CNTTNC/TEXT2GRAPH/csv_result"
if not os.path.exists(structure_data_folder_path):
  os.makedirs(structure_data_folder_path)
if not os.path.exists(csv_result_path):
  os.makedirs(csv_result_path)
````

````python
from CallAPILibrary import CallAPI
import pandas as pd
import json

api_caller = CallAPI()
# api_caller.call_api(contents=["Đếm từ 1 tới 19" for _ in range(5)], threads_num=2)
# api_caller.call_api(contents=contents, threads_num=9)
````

````python
for filename in os.listdir(prompt_folder_path):
  if filename.endswith(".json") and filename not in os.listdir(structure_data_folder_path):
    print("filename", filename)
    prompt_path = os.path.join(prompt_folder_path, filename)
    with open(prompt_path, "r") as f:
      prompts = json.load(f)
      print(prompts[0])
      threads_num_value = min(len(prompts), 9)
      api_caller.call_api(contents=prompts, threads_num=threads_num_value)
      # Đường dẫn file csv của thư viện call api
      csv_path = "./CallAPILibrary/results/results.csv"
      # Copy qua đường dẫn mới
      new_csv_path = os.path.join(csv_result_path, filename.replace(".json", ".csv"))
      os.system(f"cp {csv_path} {new_csv_path}")
      # Đọc nội dung file csv
      df = pd.read_csv(csv_path)
      # Lấy nội dung cột response
      response_column = df['response']
      # Chuyển giá trị đó thành dạng json
      response_json_list = []
      count_err = 0
      for i in range(len(response_column)):
        # kiểm tra có phải rỗng không
        if pd.isna(response_column[i]):
          continue
        response_column_data = response_column[i].replace("```json", "").replace("```", "")
        try:
          if response_column_data.endswith("]"):
            response_json = json.loads(response_column_data)
          else:
            # tìm dấu "}," cuối cùng
            last_comma_index = response_column_data.rfind("},")
            response_json = json.loads(response_column_data[:last_comma_index + 1]+ "]")
          # print("response_json", response_json)
        except:
          print("response_column_data err", response_column_data)
          count_err += 1
          continue
        response_json_list.extend(response_json)
      print("count_err", count_err)
      # Lưu vào file
      structure_data_path = os.path.join(structure_data_folder_path, filename)
      with open(structure_data_path, "w") as f:
        json.dump(response_json_list, f, indent=4, ensure_ascii=False)
    # break
````
